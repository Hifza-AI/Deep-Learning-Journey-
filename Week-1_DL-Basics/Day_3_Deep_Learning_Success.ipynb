# Difference b/w Sigmoid, ReLu and tanh activation functions

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

# Functions
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

# Plotting
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1); plt.plot(x, sigmoid); plt.title("Sigmoid")
plt.subplot(1, 3, 2); plt.plot(x, relu); plt.title("ReLU")
plt.subplot(1, 3, 3); plt.plot(x, tanh); plt.title("Tanh")
plt.show()
```

# Foundation of Weights, Bias and Activation Functions 

```python
import numpy as np
# Inputs dena 
inputs = np.array([2.0, 1.5])
# weights of these inputs 
weights = np.array([0.8, -0.5])
# Bias value 
bias = -5.0
output_layer = np.dot(weights, inputs) + bias
print("Output_Layer is: ", output_layer)
relu_output = np.maximum(0, output_layer)
sigmoid_output = 1/(1+ np.exp(-output_layer))
print("ReLu Result: ", relu_output)
print("Sigmoid Result: ", sigmoid_output)
```
