```python
import numpy as np
# Give inputs for the process
inputs = np.array([[12, 34, 78, 90],
                 [3, 2, 7, 0]
                  ])
# Provide the weights for these inputs 
weights = np.array([[0.6, 0.3, 2, 6],
                   [1,4,6,0.1]
                   ])
# Now give Bias Value for this data 
bias = np.array([8, 5]).reshape(2,1)
output_layer = np.dot (weights, inputs.T) +bias
relu_output = np.maximum(0, output_layer)
sigmoid_output = 1/(1+np.exp(-output_layer))
print("Relu Output: ", relu_output)
print("Sigmoid Output: ", sigmoid_output)
```
