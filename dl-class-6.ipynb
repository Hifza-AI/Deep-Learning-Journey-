# I implemented Backpropagation to update weights

```python
import numpy as np

# 1. Setup (Input 2 hai, target 10 hai)
x = 2.0
y_true = 50.0
w = 10  # Shuruati wazan (Weight)
lr = 0.1 # Learning Rate

# 2. Iteration (Hum 10 dafa seekhenge)
for i in range(100):
    # Forward Pass
    y_pred = w * x
    loss = (y_pred - y_true)**2
    
    # Backpropagation (Gradient nikalna)
    grad = 2 * (y_pred - y_true) * x
    
    # Weight Update
    w = w - (lr * grad)
    
    print(f"Step {i+1}: Prediction = {y_pred:.2f}, Loss = {loss:.2f}")

print(f"\nFinal Weight: {w:.2f}")
print(f"Final Prediction: {w * x:.2f}")
```
