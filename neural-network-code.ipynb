# Today I build a Neural Network Model, Train it and got Prediction from it.  

Importing Required Libraries:

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Making Model:

```python
model = Sequential()
```

Providing Data which is to be trained:

```python
X_train = np.random.rand(100, 10) 
y_train = np.random.randint(2, size=100) # 0 ya 1 target
X_test = np.random.rand(20, 10)
```

Making Hidden Layers:

```python
model.add(Dense(units = 16, activation = 'relu', input_dim = X_train.shape[1])) # 1st Layer
model.add(Dense(units = 8, activation = 'relu')) # 2nd layer
model.add(Dense(units = 1, activation = 'sigmoid')) # 3rd Layer
```

Compiling the Model:

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Training of Model starts here:

```python
model.fit(X_train, y_train, epochs = 50, batch_size = 10)
```

Final Prediction of Model:

```python
pred = model.predict(X_test)
```
